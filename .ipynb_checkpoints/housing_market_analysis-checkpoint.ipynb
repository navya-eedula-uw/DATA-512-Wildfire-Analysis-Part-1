This section focuses on visualizing the smoke estimate data and housing data, examining how they correlate, and identifying any noticeable trends. The goal is to understand the relationship between these datasets and uncover patterns that can inform further analysis.

This code creates an interactive line graph using Plotly, displaying the forecasted smoke estimates over time. The below code defines a set of datasets in a dictionary, with each dataset being visualized in its own plot. The plots show how the data changes over time by plotting the values against the date. For each dataset, the plot includes a line graph with markers, allowing the user to visually track changes over time. This code processes the `DATE` column in four datasets, ensuring that the dates are in a consistent format that the system can recognize as actual date values. I then creates a new column called `Year` by extracting the year from the `DATE` column. This simplifies year-based analysis, allowing you to group or filter the data by year in later steps. By standardizing the date format and adding a year column, the code makes the data more uniform and easier to work with.

````python
# Convert the 'DATE' column to datetime format for all datasets
annual_population_data['DATE'] = pd.to_datetime(annual_population_data['DATE'])  # Standardize 'DATE' in population data
monthly_active_home_listings['DATE'] = pd.to_datetime(monthly_active_home_listings['DATE'])  # Standardize 'DATE' in active home listings
monthly_median_price_per_square_foot['DATE'] = pd.to_datetime(monthly_median_price_per_square_foot['DATE'])  # Standardize 'DATE' in price data
monthly_new_home_construction_permits['DATE'] = pd.to_datetime(monthly_new_home_construction_permits['DATE'])  # Standardize 'DATE' in home permits

# Extract the 'Year' from the 'DATE' column for all datasets
annual_population_data['Year'] = annual_population_data['DATE'].dt.year  # Extract year for population data
monthly_active_home_listings['Year'] = monthly_active_home_listings['DATE'].dt.year  # Extract year for active home listings
monthly_median_price_per_square_foot['Year'] = monthly_median_price_per_square_foot['DATE'].dt.year  # Extract year for price data
monthly_new_home_construction_permits['Year'] = monthly_new_home_construction_permits['DATE'].dt.year  # Extract year for home permits
````

````python
# Make a copy of smoke_estimate_data to work on annual data analysis
annual_df = smoke_estimate_data.copy()

# Merge the annual population data into monthly datasets for active home listings and construction permits
monthly_active_home_listings = pd.merge(
    monthly_active_home_listings, 
    annual_population_data[['Year', 'annual_population_data']], 
    on='Year', 
    how='left'
)
monthly_new_home_construction_permits = pd.merge(
    monthly_new_home_construction_permits, 
    annual_population_data[['Year', 'annual_population_data']], 
    on='Year', 
    how='left'
)

# Normalize monthly data by dividing key fields by annual population to adjust for population changes
monthly_active_home_listings['normalized_listings'] = (
    monthly_active_home_listings['monthly_active_home_listings'] / 
    monthly_active_home_listings['annual_population_data_x']
)
monthly_new_home_construction_permits['normalized_permits'] = (
    monthly_new_home_construction_permits['monthly_new_home_construction_permits'] / 
    monthly_new_home_construction_permits['annual_population_data_x']
)

# Aggregate normalized data to the annual level
annual_listings = monthly_active_home_listings.groupby('Year')['normalized_listings'].mean().reset_index()
annual_permits = monthly_new_home_construction_permits.groupby('Year')['normalized_permits'].sum().reset_index()
annual_price = monthly_median_price_per_square_foot.groupby('Year')['monthly_median_price_per_square_foot'].mean().reset_index()

# Combine the annual summaries into the copied smoke estimate dataset
annual_df = pd.merge(annual_df, annual_listings, on='Year', how='left')
annual_df = pd.merge(annual_df, annual_permits, on='Year', how='left')
annual_df = pd.merge(annual_df, annual_price, on='Year', how='left')
````

# Normalizing Housing Data using Min-Max Scalar
When analyzing housing and construction data, normalizing metrics by population can provide a clearer understanding of trends and patterns. This section outlines the rationale behind normalizing or not normalizing key metrics like active home listings, median price per square foot, and construction permits. By applying this approach, we can differentiate changes driven by population growth from those influenced by market dynamics or economic conditions. Below is the reasoning for each metric:

1. **monthly_active_home_listings**  
I decided to normalize this field by annual population in the Denver-Aurora-Lakewood region. As the population grows, the number of active home listings is likely to increase proportionally. Normalizing this metric allows for fair comparison across different time periods or between cities of different sizes. It helps distinguish between increases in listings due to population growth versus market conditions. 


2. **monthly_median_price_per_square_foot**  
This field does not need to be normalized by annual population. It's already a relative measure (price per unit area) that accounts for differences in property sizes.


3. **monthly_new_home_construction_permits**  
I decided to normalize this field by annual population in the Denver-Aurora-Lakewood region. As the population grows, the demand for new housing typically increases, leading to more construction permits. Normalizing this metric allows for a more accurate assessment of construction activity relative to population size. It helps distinguish between increases in construction due to population growth versus other economic factors.

To standardise my data and compare all these different variables accross different scales and ranges. I used the min-max scaling technique. It brings all features to the same scale (0-1), making comparisons easier. Furthuremore, plotting multiple variables on the same graph becomes more meaningful. It preserves zero values and maintains the relative distances between data points. I think that it becomes easier to identify which variables are showing the most significant changes over time.The scaled values are easily interpretable as percentages of the range. I believe that for my model, the features on similar scales can improve performance and convergence.  

**Formula:**  

The formula for Min-Max scaling is:

$$X_{scaled} = \frac{X - X_{min}}{X_{max} - X_{min}}$$

Where:  
- $X$ is the original value
- $X_{min}$ is the minimum value of the feature
- $X_{max}$ is the maximum value of the feature  

````python
# Initialize MinMaxScaler
scaler = MinMaxScaler()

annual_df['normalized_listings'] = scaler.fit_transform(annual_df[['normalized_listings']])
annual_df['normalized_permits'] = scaler.fit_transform(annual_df[['normalized_permits']])
annual_df['monthly_median_price_per_square_foot'] = scaler.fit_transform(annual_df[['monthly_median_price_per_square_foot']])
````

# Data Preparation and Preprocessing Pipeline for Annual Dataset

This code processes multiple datasets related to housing data and prepares them for further analysis. The key tasks it performs include data aggregation, handling missing values, and normalization of certain columns. Here's a breakdown of the steps:

1. **Aggregation to Annual Data**: The `aggregate_monthly_to_annual()` function converts monthly data into yearly data by calculating the average value for each year. It is applied to several datasets like housing listings and home construction permits.

2. **Backfilling Missing Values**: The `backfill_with_linear_model()` function fills in missing data for specific columns (such as smoke estimates and housing data) using a linear regression model. It predicts missing values based on the known yearly data, allowing for a smooth continuation of data where gaps exist. This was used for the variables where the datsa did not date all the way back to 1964 from where the smoke estimate data begins.

3. **Merging Datasets**: The `merge_datasets()` function combines multiple datasets—such as smoke estimates, population data, housing listings, and construction permits—into a single DataFrame. This is done by merging on the year, ensuring each dataset is aligned by the year for analysis.

4. **Min-Max Scaling**: The `min_max_scale_columns()` function normalizes certain columns using Min-Max scaling, transforming the data to a 0-1 range. This is helpful to standardize the scale of different features for further modeling or comparison.

5. **Final Data Preparation**: The pipeline aggregates the datasets, fills missing values, scales relevant columns, and selects the most important columns for the final dataset, which will be used for analysis.


````python
# Data Preparation Functions

def aggregate_monthly_to_annual(monthly_df, value_col):
    """
    Aggregates monthly data to annual by taking the average value for each year.
    
    Args:
        monthly_df (pd.DataFrame): DataFrame with monthly data.
        value_col (str): The column name to aggregate (e.g., 'monthly_active_home_listings').
    
    Returns:
        pd.DataFrame: DataFrame with annual data, where values are averaged by year.
    """
    monthly_df = monthly_df.copy()  # Create a copy to avoid modifying the original DataFrame
    monthly_df['Year'] = pd.to_datetime(monthly_df['DATE']).dt.year  # Extract year from the 'DATE' column
    annual_df = monthly_df.groupby('Year')[value_col].mean().reset_index()  # Group by year and calculate the average
    return annual_df

def backfill_with_linear_model(data, target_col, start_year, end_year):
    """
    Backfills missing values in the target column using a linear regression model.
    
    Args:
        data (pd.DataFrame): DataFrame with missing values.
        target_col (str): The column to fill missing values for (e.g., 'Forecasted_Smoke_Estimate').
        start_year (int): The starting year for the dataset.
        end_year (int): The ending year for the dataset.
    
    Returns:
        pd.DataFrame: DataFrame with missing values in target_col backfilled using linear regression.
    
    Raises:
        ValueError: If there are no known data points to fit the regression model.
    """
    # Create a full range of years from start_year to end_year
    full_years = pd.DataFrame({'Year': range(start_year, end_year + 1)})
    data = pd.merge(full_years, data, on='Year', how='left')  # Merge full year range with the data
    
    # Known data points to train the model (drop rows with missing target_col values)
    known_data = data.dropna(subset=[target_col])
    if known_data.empty:
        raise ValueError(f"No known data points to fit regression for {target_col}.")
    
    # Fit a linear regression model on known data
    model = LinearRegression()
    model.fit(known_data[['Year']], known_data[target_col])
    
    # Predict missing values using the fitted model
    missing_data = data[data[target_col].isnull()]
    if not missing_data.empty:
        data.loc[missing_data.index, target_col] = model.predict(missing_data[['Year']])
    
    return data

def merge_datasets(smoke_df, annual_population_data, monthly_active_home_listings, monthly_median_price_per_square_foot, monthly_new_home_construction_permits):
    """
    Merges multiple datasets (smoke, population, housing, etc.) into a single annual dataset.
    
    Args:
        smoke_df (pd.DataFrame): Smoke estimates dataset.
        annual_population_data (pd.DataFrame): Population data per year.
        monthly_active_home_listings (pd.DataFrame): Monthly housing listings data.
        monthly_median_price_per_square_foot (pd.DataFrame): Monthly median price data.
        monthly_new_home_construction_permits (pd.DataFrame): Monthly new home construction data.
    
    Returns:
        pd.DataFrame: A single DataFrame with all datasets merged by year.
    """
    # Merge smoke estimates with annual population data
    annual_df = pd.merge(smoke_df, annual_population_data, on='Year', how='outer')
    
    # Aggregate monthly data to annual for housing listings, median prices, and new permits
    monthly_active_home_listings = aggregate_monthly_to_annual(monthly_active_home_listings, 'monthly_active_home_listings')
    annual_df = pd.merge(annual_df, monthly_active_home_listings, on='Year', how='outer')
    
    # Add median price per square foot
    monthly_median_price_per_square_foot = aggregate_monthly_to_annual(monthly_median_price_per_square_foot, 'monthly_median_price_per_square_foot')
    annual_df = pd.merge(annual_df, monthly_median_price_per_square_foot, on='Year', how='outer')
    
    # Add new home construction permits
    monthly_new_home_construction_permits = aggregate_monthly_to_annual(monthly_new_home_construction_permits, 'monthly_new_home_construction_permits')
    annual_df = pd.merge(annual_df, monthly_new_home_construction_permits, on='Year', how='outer')
    
    # Normalize listings and permits by population size
    monthly_active_home_listings['normalized_listings'] = monthly_active_home_listings['monthly_active_home_listings'] / annual_df['annual_population_data']
    monthly_new_home_construction_permits['normalized_permits'] = monthly_new_home_construction_permits['monthly_new_home_construction_permits'] / annual_df['annual_population_data']
    
    return annual_df

def min_max_scale_columns(df, columns):
    """
    Applies Min-Max scaling to specified columns of the DataFrame, transforming values to the range [0, 1].
    
    Args:
        df (pd.DataFrame): The input DataFrame to scale.
        columns (list): List of column names to apply scaling on.
    
    Returns:
        pd.DataFrame: The DataFrame with the specified columns scaled.
    """
    df = df.copy()  # To avoid changing the original DataFrame
    for col in columns:
        min_val = df[col].min()  # Find the minimum value of the column
        max_val = df[col].max()  # Find the maximum value of the column
        if min_val == max_val:  # Handle columns with no variation
            df[col] = 0.5  # Assign mid-value for constant columns
        else:
            df[col] = (df[col] - min_val) / (max_val - min_val)  # Apply Min-Max scaling
    return df
````

````python
# Full Pipeline
start_year = 1964  # Define the start year for the dataset
end_year = 2020    # Define the end year for the dataset

# Step 1: Merge datasets into a single DataFrame
annual_df = merge_datasets(smoke_estimate_data, annual_population_data, monthly_active_home_listings, monthly_median_price_per_square_foot, monthly_new_home_construction_permits)

# Step 2: Backfill missing values using linear regression for each relevant column
for col in ['Forecasted_Smoke_Estimate', 'monthly_active_home_listings', 
            'monthly_median_price_per_square_foot', 'monthly_new_home_construction_permits']:
    annual_df = backfill_with_linear_model(annual_df, col, start_year, end_year)

# Step 3: Apply Min-Max scaling to specified columns
annual_df = min_max_scale_columns(annual_df, [
    'monthly_active_home_listings',
    'monthly_median_price_per_square_foot',
    'monthly_new_home_construction_permits'
])

# Step 4: Select and arrange relevant columns for the final dataset
annual_df = annual_df[[
    'Year',  # Include the year for reference
    'Forecasted_Smoke_Estimate',  # Include smoke estimate data
    'DATE',  # Include date reference for consistency
    'monthly_active_home_listings',  # Include housing data
    'monthly_median_price_per_square_foot',  # Include housing price data
    'monthly_new_home_construction_permits'  # Include construction permits data
]]

# Display the first few rows of the final DataFrame
annual_df.head()  # Show the resulting DataFrame to check the final data
````

# Time Series Forecasting with SARIMA Model and Impact Analysis Based on Smoke Scenarios

This section is focused on analyzing time series data using SARIMA models with an exogenous variable, specifically the smoke estimate. 

1. **Preparing the Data**: The first step is to prepare the data. The `prepare_data()` function ensures the 'Year' column is properly converted to a datetime index and handles missing values using linear interpolation and forward/backward filling. 

2. **Building the SARIMA Model**: The `build_sarima_model()` function splits the data into training and test sets. It then fits a SARIMA model to the training data, where the target variable (e.g., home listings, prices, etc.) is modeled with an external factor like a smoke estimate. 

3. **Forecasting Scenarios**: The `forecast_scenarios()` function generates two forecast scenarios: one where the smoke estimate increases over time and another where it stays constant. It then generates forecasts for both scenarios using the SARIMA model.

4. **Plotting the Forecasts**: The `plot_forecasts()` function plots both the historical data and the two forecast scenarios. The historical data is shown up until the training year, and the forecasts are plotted for future years, allowing us to compare the two scenarios visually.

````python
def prepare_data(df):
    """
    Prepare data for time series analysis.
    
    This function prepares the input dataframe by setting the 'Year' column as the datetime index, 
    handling missing values through interpolation, and using forward/backward filling methods for any
    remaining missing values at the edges.
    """
    # Convert Year to proper datetime index
    df = df.copy()  # Create a copy of the dataframe to avoid modifying the original
    df.index = pd.to_datetime(df['Year'].astype(str), format='%Y')  # Convert 'Year' column to datetime and set as index
    
    # Handle missing values
    for column in df.columns:
        if df[column].isnull().any():  # Check if there are any missing values in the column
            # Use linear interpolation first
            df[column] = df[column].interpolate(method='linear')  # Fill missing values with linear interpolation
            # If any remaining NaN at the edges, use forward/backward fill
            df[column] = df[column].fillna(method='ffill').fillna(method='bfill')  # Forward fill or backward fill
            
    return df
````

The ARIMAX model used here is quite similar to the one I employed in Part 1 for smoke estimate forecasting, with the only difference being the model parameters.

The ARIMAX (AutoRegressive Integrated Moving Average with eXogenous variables) model is an extension of the ARIMA model that incorporates exogenous variables—external factors that can influence the target variable. It is useful for time series forecasting when additional variables, beyond the target itself, affect the predictions.
The ARIMAX model consists of several components: the autoregressive (AR) part, which uses past values of the target variable to predict future values; the moving average (MA) part, which uses past errors to adjust the forecast; and the integrated (I) part, which handles non-stationary data by differencing. 

In this model, the function `build_sarima_model` splits the dataset into training and test sets based on a specified year (train_end_year). It then applies the ARIMAX model to the training data, using the target variable (e.g., home listings, prices) as the dependent variable and the smoke estimate as an exogenous factor. The SARIMA model's order is set with (1, 1, 1) for the AR, I, and MA terms, and a seasonal order of (1, 1, 1, 12) is specified to account for yearly seasonality. Since it is annual data, we do not account for seasonality. The function fits the model to the training data and returns the model results along with the training and test sets for further evaluation. By incorporating the smoke estimate as an exogenous variable, the model can predict the target variable while accounting for the impact of smoke levels, providing a more accurate and context-aware forecast.

````python
def build_sarima_model(data, target_col, smoke_col, train_end_year=2020):
    """
    Build SARIMA model with exogenous variable (smoke estimate).
    
    This function splits the data into training and test sets, fits a SARIMA model with the given target
    variable and an exogenous smoke estimate, and returns the model results along with the training and test sets.
    """
    # Split data into train and test
    train = data[data.index.year <= train_end_year]  # Training set: data until the specified year
    test = data[data.index.year > train_end_year]  # Test set: data after the specified year
    
    # Fit SARIMA model
    model = SARIMAX(train[target_col],  # Target column for SARIMA model
                    exog=train[smoke_col],  # Exogenous variable (smoke estimate)
                    order=(1, 1, 1),  # ARIMA model order
                    seasonal_order=(1, 1, 1, 12))  # Seasonal components of the SARIMA model
    
    results = model.fit()  # Fit the SARIMA model
    
    return results, train, test
````

````python
def forecast_scenarios(model, train_data, future_periods, smoke_col):
    """
    Generate forecasts for both increasing and constant smoke scenarios.
    
    This function generates two forecast scenarios based on the smoke estimate: 
    one where smoke increases over time and one where it remains constant. 
    It returns the forecasts for both scenarios along with the future smoke values.
    """
    # Create future smoke scenarios
    last_smoke = train_data[smoke_col].iloc[-1]  # Last observed smoke level
    smoke_trend = (train_data[smoke_col].iloc[-1] - train_data[smoke_col].iloc[0]) / len(train_data)  # Smoke trend
    
    # Increasing scenario
    future_smoke_increasing = np.array([last_smoke + smoke_trend * i for i in range(1, future_periods + 1)])
    
    # Constant scenario
    future_smoke_constant = np.array([last_smoke] * future_periods)
    
    # Generate forecasts
    forecast_increasing = model.get_forecast(steps=future_periods, exog=future_smoke_increasing)
    forecast_constant = model.forecast(future_periods, exog=future_smoke_constant)
    
    # Extract confidence intervals for increasing scenario
    conf_int_increasing = forecast_increasing.conf_int(alpha=0.05)
    forecast_mean_increasing = forecast_increasing.predicted_mean
    
    return forecast_mean_increasing, forecast_constant, conf_int_increasing, future_smoke_increasing, future_smoke_constant
````

````python
def plot_forecasts(data, target_col, forecast_inc, conf_int_inc, forecast_const, train_end_year):
    """
    Plot the forecasts against actual data.
    
    This function plots the actual data up to the training year and the forecasts for both increasing and constant
    smoke scenarios after the training year. It includes confidence intervals for the increasing smoke scenario.
    """
    plt.figure(figsize=(12, 6))  # Set up the figure for plotting
    
    # Plot historical data
    plt.plot(data.index[data.index.year <= train_end_year], 
             data[target_col][data.index.year <= train_end_year], 
             'b-', label='Historical Data')  # Plot historical data in blue
    
    # Plot forecasts
    future_dates = pd.date_range(start=str(train_end_year+1), 
                                 periods=len(forecast_inc), 
                                 freq='Y')  # Generate future dates for plotting
    
    # Forecast lines for increasing and constant smoke
    plt.plot(future_dates, forecast_inc, 'r--', label='Forecast (Increasing Smoke)')  # Forecast for increasing smoke
    plt.plot(future_dates, forecast_const, 'g--', label='Forecast (Constant Smoke)')  # Forecast for constant smoke
    
    # Confidence intervals for increasing smoke
    plt.fill_between(future_dates, 
                     conf_int_inc.iloc[:, 0], 
                     conf_int_inc.iloc[:, 1], 
                     color='red', alpha=0.2)  # Confidence interval
    
    # Adding labels and legend
    plt.title(f'Forecasts for {target_col}')
    plt.xlabel('Year')
    plt.ylabel('Value')
    plt.legend(loc='best')  # Explicitly call legend and set the location
    plt.grid(True)
    plt.show()
````

At the end of this code, I've attempted to do a simple impact analysis.

The impact analysis in this sub-section evaluates the effect of different smoke scenarios on the target variables. Specifically, it compares two forecast scenarios for each target variable: one where smoke levels are projected to increase and one where smoke levels remain constant. The goal is to understand how changes in smoke levels (increasing or constant) can influence the forecasted values of variables like home listings, prices, and construction permits.

For each target variable, the code calculates the average predicted impact by taking the difference between the forecast under the increasing smoke scenario and the forecast under the constant smoke scenario, dividing it by the forecast under the constant smoke scenario, and then multiplying by 100 to express the impact as a percentage. This gives an idea of how much the target variable changes on average when smoke levels are expected to increase. Additionally, the maximum predicted impact is calculated, which identifies the largest difference (as a percentage) between the two scenarios, providing insight into the extreme possible effects of increasing smoke levels. These metrics are printed for each target variable, allowing for a clear comparison of how different smoke trends might influence the outcomes being forecasted. 

Here are the imact analysis formulas in equation format -

#### Average Predicted Impact Formula:
The **average predicted impact** is calculated as:

$$
\text{Average Impact} = \frac{{\text{forecast_increasing} - \text{forecast_constant}}}{{\text{forecast_constant}}} \times 100
$$

Where:
- `forecast_increasing` is the forecast under the increasing smoke scenario.
- `forecast_constant` is the forecast under the constant smoke scenario.

The formula computes the percentage difference between the two forecasts and averages the result across all future periods.

#### Maximum Predicted Impact Formula:
The **maximum predicted impact** is calculated as:

$$
\text{Maximum Impact} = \max \left( \frac{{\text{forecast_increasing} - \text{forecast_constant}}}{{\text{forecast_constant}}} \times 100 \right)
$$

Where:
- The maximum function (`max`) is applied to find the largest percentage difference between the forecasts for increasing and constant smoke scenarios.

These formulas quantify the impact of smoke level changes on the target variables, giving both an average and the most extreme change across the forecast period.

````python
# Prepare data
data = prepare_data(annual_df)  # Preprocess data for time series analysis

# Variables to analyze
target_variables = ['monthly_active_home_listings',
                    'monthly_median_price_per_square_foot',
                    'monthly_new_home_construction_permits']

# Future periods to forecast
future_periods = 30  # Forecast from 2024 to 2050

# Results dictionary to store forecasts
results = {}

# 1. Individual SARIMA models for each target
for target in target_variables:
    print(f"\nAnalyzing {target}")

    # Build SARIMA model
    model, train, test = build_sarima_model(
        data,
        target,
        'Forecasted_Smoke_Estimate',
        train_end_year=2020
    )

    # Generate forecasts
    forecast_inc, forecast_const, conf_int_inc, smoke_inc, smoke_const = forecast_scenarios(
        model,
        train,
        future_periods,
        'Forecasted_Smoke_Estimate'
    )

    # Store results
    results[target] = {
        'forecast_increasing': forecast_inc,
        'forecast_constant': forecast_const,
        'confidence_interval_increasing': conf_int_inc,
        'actual_data': data[target]
    }

    # Plot results
    plot_forecasts(data, target, forecast_inc, conf_int_inc, forecast_const, 2020)

# Print summary statistics
print("\nImpact Analysis:")
for target in target_variables:
    inc_forecast = results[target]['forecast_increasing']
    const_forecast = results[target]['forecast_constant']

    # Calculate average impact
    avg_impact = ((inc_forecast - const_forecast) / const_forecast * 100).mean()  # Calculate average impact

    print(f"\n{target}:")
    print(f"Average predicted impact: {avg_impact:.2f}%")
    print(f"Maximum predicted impact: {((inc_forecast - const_forecast) / const_forecast * 100).max():.2f}%")
````

### Time Series Forecasting: Comparing the Three Different Variables

#### Active Home Listings
The first graph presents the trend of active home listings over time, with historical data depicted in blue and two forecasted trends shown in dashed lines.
````python
import pandas as pd
import numpy as np
from statsmodels.tsa.statespace.sarimax import SARIMAX
import matplotlib.pyplot as plt

def prepare_monthly_data(df):
    df = df.copy()
    df['DATE'] = pd.to_datetime(df['DATE'])
    df.set_index('DATE', inplace=True)
    
    for column in df.columns:
        if df[column].isnull().any():
            df[column] = df[column].interpolate(method='linear')
            df[column] = df[column].fillna(method='ffill').fillna(method='bfill')
    
    return df

def split_wildfire_nonwildfire(df):
    # Assuming wildfire months are June through September
    wildfire_months = df[df.index.month.isin([6, 7, 8, 9])]
    nonwildfire_months = df[~df.index.month.isin([6, 7, 8, 9])]
    return wildfire_months, nonwildfire_months

def build_sarima_model(data, target_col, smoke_col, train_end_date='2020-12-31'):
    train = data[data.index <= train_end_date]
    test = data[data.index > train_end_date]
    
    model = SARIMAX(train[target_col],
                    exog=train[smoke_col],
                    order=(2, 1, 2),
                    seasonal_order=(1, 1, 1, 12))
    
    results = model.fit()
    
    return results, train, test

def forecast_scenarios(model, train_data, future_periods, smoke_col):
    last_smoke = train_data[smoke_col].iloc[-1]
    smoke_trend = (train_data[smoke_col].iloc[-1] - train_data[smoke_col].iloc[0]) / len(train_data)
    
    future_smoke_increasing = np.array([last_smoke + smoke_trend * i for i in range(1, future_periods + 1)])
    future_smoke_constant = np.array([last_smoke] * future_periods)
    
    forecast_increasing = model.forecast(future_periods, exog=future_smoke_increasing)
    forecast_constant = model.forecast(future_periods, exog=future_smoke_constant)
    
    return forecast_increasing, forecast_constant, future_smoke_increasing, future_smoke_constant

def plot_forecasts(data, target_col, forecast_inc, forecast_const, train_end_date, title_suffix):
    plt.figure(figsize=(12, 6))
    
    plt.plot(data.index[data.index <= train_end_date], 
             data[target_col][data.index <= train_end_date], 
             'b-', label='Historical Data')
    
    future_dates = pd.date_range(start=pd.to_datetime(train_end_date) + pd.Timedelta(days=1), 
                                periods=len(forecast_inc), 
                                freq='M')
    
    plt.plot(future_dates, forecast_inc, 'r--', label='Forecast (Increasing Smoke)')
    plt.plot(future_dates, forecast_const, 'g--', label='Forecast (Constant Smoke)')
    
    plt.title(f'Forecasts for {target_col} - {title_suffix}')
    plt.xlabel('Date')
    plt.ylabel('Value')
    plt.legend()
    plt.grid(True)
    plt.show()

def main():
    # Prepare data
    data = prepare_monthly_data(monthly_df)
    
    # Split into wildfire and non-wildfire months
    wildfire_data, nonwildfire_data = split_wildfire_nonwildfire(data)
    
    target_variables = ['monthly_new_home_construction_permits']
    
    future_periods = 12 * 30  # Monthly data from 2021 to 2050
    
    results = {}
    
    for dataset, dataset_name in [(wildfire_data, "Wildfire Months"), (nonwildfire_data, "Non-Wildfire Months")]:
        results[dataset_name] = {}
        
        for target in target_variables:
            print(f"\nAnalyzing {target} for {dataset_name}")
            
            model, train, test = build_sarima_model(
                dataset,
                target,
                'Forecasted_Smoke_Estimate',
                train_end_date='2020-12-31'
            )
            
            forecast_inc, forecast_const, smoke_inc, smoke_const = forecast_scenarios(
                model,
                train,
                future_periods,
                'Forecasted_Smoke_Estimate'
            )
            
            results[dataset_name][target] = {
                'forecast_increasing': forecast_inc,
                'forecast_constant': forecast_const,
                'actual_data': dataset[target]
            }
            
            plot_forecasts(dataset, target, forecast_inc, forecast_const, '2020-12-31', dataset_name)
        
        print(f"\nImpact Analysis for {dataset_name}:")
        for target in target_variables:
            inc_forecast = results[dataset_name][target]['forecast_increasing']
            const_forecast = results[dataset_name][target]['forecast_constant']
            
            avg_impact = ((inc_forecast - const_forecast) / const_forecast * 100).mean()
            
            print(f"\n{target}:")
            print(f"Average predicted impact: {avg_impact:.2f}%")
            print(f"Maximum predicted impact: {((inc_forecast - const_forecast) / const_forecast * 100).max():.2f}%")

    return results

if __name__ == "__main__":
    results = main()
````
